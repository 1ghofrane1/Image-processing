```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage import color, data, exposure

# Load the 'cell' and 'coffee' images
cell = data.cell()
coffee = data.coffee()

# Resize the 'coffee' image to match the shape of the 'cell' image
coffee_resized = skimage.transform.resize(coffee, cell.shape)

# Convert the 'cell' image to a color image
cell_color = color.gray2rgb(cell)

# Multiply the 'cell' and 'coffee' images
mult_image = cell_color * coffee_resized

# Display the original and processed images using Matplotlib's 'imshow' function
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(10, 5))

ax0.imshow(cell, cmap='gray')
ax0.set_title('Cell')

ax1.imshow(coffee_resized, cmap='gray')
ax1.set_title('Coffee Resized')

ax2.imshow(mult_image)
ax2.set_title('Cell x Coffee')

plt.show()
```

# Notebook 2 : filtres lin√©aires 
    (filtres par convolution)

```python
import skimage
import skimage.filters
import matplotlib.pyplot as plt
import numpy as np
```

```python
I= skimage.data.camera()
plt.imshow(I, 'gray')
plt.show()
```

```python
from scipy.ndimage import uniform_filter
I_moy= uniform_filter(I, size=(150,150))
plt.imshow(I_moy,'gray')
plt.show()
```

```python
from scipy.ndimage import median_filter
I_moy= uniform_filter(I, size=(150,150))
plt.imshow(I_moy,'gray')
plt.show()
```

```python
I_gauss= skimage.filters.gaussian(I,15)
plt.imshow(I_gauss, 'gray')
plt.show()
```

```python
I_sobel= skimage.filters.sobel(I)
# on calcule la norme de gradient avec cette fonction
plt.imshow(I_sobel, 'gray')
plt.show()
```

```python
I_scharr= skimage.filters.scharr(I)
# on calcule la norme de gradient avec cette fonction
plt.imshow(I_scharr, 'gray')
plt.show()
```

```python
I_hessian = skimage.filters.hessian(I, mode='constant')
plt.imshow(I_hessian, 'gray')
plt.show()
```
